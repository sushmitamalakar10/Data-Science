# indexing
 accessing individual elements using loc and iloc

### indexing using loc
 loc means location... we can indexx elements using label based indexing

obj.loc[]

```python
df1.loc[1, 'A']
```

```python
df1.loc[3, 'C']
```

### iloc
iloc: integer based location
- obj.iloc(row_index, col_index)

```python
df1.iloc[2,1]
```

```python
df1.iloc[-1,-1]
```

### Slicing in pandas
- using loc
- obj.loc[row_level_slicing, col_level_slicing]

```python
df1.loc[0:1, :'col2']
```

```python
df1.loc[:1, :'col2']
```

### slicing using iloc
- obj.iloc[row_level_slicing, col_level_slicing]

```python
df1.iloc[:2, -2:]
```
