### Joins in pandas
#### types of joins: inner, outer, left, right

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({
    "name" : ['sunil', 'anil', 'ram', 'shyam', 'sita', 'gita'],
    "age" : [20,21,24,26,45,34],
    "city" : ['ktm', 'pkr', 'htd', 'bkt', 'brt', 'palpa'],
    "id" : [1,2,3,4,5,6]
})
df1
```

```python
df2 = pd.DataFrame({
    "fruits_name": ['apple', 'mango', 'banana', 'mango', 'grapes', 'kiwi', 'litchi', 'papaya', 'berry'], 
    "id" : [1,2,3,7,8,9,10,11, 12]
})
df2
```

### Inner join

```python
pd.merge(df1, df2, how='inner', on='id')
```

### Left join

```python
pd.merge(df1, df2, how="left", on="id")
```

```python
pd.merge(df1, df2, how="right", on="id")
```

```python
pd.merge(df1, df2, how="outer", on="id")
```

### Concatenation in Pandas

```python
data1 = pd.DataFrame(np.arange(1, 13).reshape(4,3), columns=['col1', 'col2', 'col3'])
data1
```

```python
data2 = pd.DataFrame(np.arange(1, 25).reshape(6,4), columns=['col1', 'col2', 'col3','col4'])
data2
```

```python
pd.concat([data1, data2], axis=1)
```

```python
pd.concat([data1, data2], axis=0)
```
